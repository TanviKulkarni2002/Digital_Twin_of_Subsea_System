```python
# Dropping unimportant features
df=df.drop(['id','oilpump_power','gaccx','gaccy','gaccz','haccx','haccy','haccz','acmotor'],axis=1)
```

```python
# Scaling
import pandas as pd
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
scaled_data = pd.DataFrame(scaled_data, columns=df.columns)
scaled_data.head()
```

```python
# Target variable: outlet_pressure_bar
from sklearn.model_selection import train_test_split
X = scaled_data.drop(['outlet_pressure_bar'],axis=1)
y = scaled_data['outlet_pressure_bar']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

y_train =  y_train.to_numpy().reshape(-1, 1)
y_test = y_test.to_numpy().reshape(-1, 1)

# Print the shapes of the training and testing sets
print("Training set shape:", X_train.shape, y_train.shape)
print("Testing set shape:", X_test.shape, y_test.shape)
```

```python
# Building the model
from sklearn.datasets import make_regression
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
model = AdaBoostRegressor(base_estimator=DecisionTreeRegressor(max_depth=8,min_samples_split=6,min_samples_leaf=1, max_features=16),n_estimators=400, learning_rate=0.01, random_state=42)
model.fit(X_train, y_train)
```

```python
# Predicting on test data
predictions = model.predict(X_test)
```

```python
# Calculating RMSE
import numpy as np
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, predictions)
rmse=np.sqrt(mse)
print("RMSE:", rmse)
```
