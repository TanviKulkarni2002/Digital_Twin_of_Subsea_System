```python
#run
# Scaling
import pandas as pd
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
scaled_X_train = pd.DataFrame(scaled_X_train, columns=X.columns)
scaled_X_test = pd.DataFrame(scaled_X_test, columns=X.columns)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

poly_reg = PolynomialFeatures(degree=2,include_bias=False)
X_train_poly = poly_reg.fit_transform(scaled_X_train)
X_test_poly = poly_reg.fit_transform(scaled_X_test)

# poly_reg.fit(X_train_poly, y_train)
model = LinearRegression()
model.fit(X_train_poly, y_train)
```

```python
# pred = np.array()
# lin2.predict(poly.fit_transform(pred2array))
predictions = model.predict(X_test_poly)
```

```python
# Calculating RMSE
import numpy as np
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, predictions)
rmse=np.sqrt(mse)
print("RMSE:", rmse)
```
